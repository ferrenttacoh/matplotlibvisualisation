```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/PBO/NilaiKimdas.csv")
df.dropna(axis=0)
```

```python
#Line Chart
f = plt.figure()
myaxes = f.add_axes([4.0,4.0,2.0,2.0])

myaxes.plot(df['Nama'], df['Nilai 1'], color='navy', lw=3, label = 'Nilai 1')
myaxes.plot(df['Nama'], df['Nilai 2'], color='magenta', lw=3, label = 'Nilai 2')
myaxes.plot(df['Nama'], df['Nilai 3'], color='cyan', lw=3, label = 'Nilai 3')
myaxes.plot(df['Nama'], df['Nilai 4'], color='orange', lw=3, label = 'Nilai 4')

myaxes.grid(True)
myaxes.legend()
myaxes.set_ylabel('NILAI')
myaxes.set_xlabel('NAMA BELAKANG')
myaxes.set_title('NILAI MATAKULIAH KIMIA DASAR')
```

```python
#Scatter Plot
f = plt.figure()
myaxes = f.add_axes([4.0,4.0,2.0,2.0])

myaxes.scatter(df['Nama'], df['Nilai 1'], color='navy', lw=3, label = 'Nilai 1')
myaxes.scatter(df['Nama'], df['Nilai 2'], color='magenta', lw=3, label = 'Nilai 2')
myaxes.scatter(df['Nama'], df['Nilai 3'], color='cyan', lw=3, label = 'Nilai 3')
myaxes.scatter(df['Nama'], df['Nilai 4'], color='orange', lw=3, label = 'Nilai 4')
myaxes.grid(True)

myaxes.legend()
myaxes.set_ylabel('NILAI')
myaxes.set_xlabel('NAMA BELAKANG')
myaxes.set_title('NILAI MATAKULIAH KIMIA DASAR')
```

```python
#Bar Chart
barWidth = 0.2
fig = plt.subplots(figsize =(20, 10))
 
# set height of bar
Nilai_1 = df['Nilai 1']
Nilai_2 = df['Nilai 2']
Nilai_3 = df['Nilai 3']
Nilai_4 = df['Nilai 4']
 
# Set position of bar on X axis
bar1 = np.arange(len(Nilai_1))
bar2 = [x + barWidth for x in bar1]
bar3 = [x + barWidth for x in bar2]
bar4 = [x + barWidth for x in bar3]
 
# Make the plot
plt.bar(bar1, Nilai_1, color ='navy', width = barWidth,
        edgecolor ='grey', label ='Nilai 1')
plt.bar(bar2, Nilai_2, color ='magenta', width = barWidth,
        edgecolor ='grey', label ='Nilai 2')
plt.bar(bar3, Nilai_3, color ='cyan', width = barWidth,
        edgecolor ='grey', label ='Nilai 3')
plt.bar(bar4, Nilai_4, color ='orange', width = barWidth,
        edgecolor ='grey', label ='Nilai 4')
 
# Adding Xticks
plt.grid(True)
plt.xlabel('Nama Belakang', fontweight ='bold', fontsize = 20)
plt.ylabel('Nilai', fontweight ='bold', fontsize = 20)
plt.xticks([r + barWidth for r in range(len(Nilai_1))],df['Nama'])
 
plt.legend()
plt.show()
```
